```python
import pandas
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype'] = 'none'
```

```python
directory = "./thauera-files/ThaueraPaper/"
```

```python
colors = ['#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f','#ff7f00','#cab2d6','#6a3d9a','#ffff99','#b15928']
```

```python
len(colors)
```

## Aerobic

```python
Carbon = pandas.read_csv("{}/carbonLethalAer.csv".format(directory),header=None,index_col=0)[1].sort_values()
Enzymes = pandas.read_csv("{}/enzymesLethalAer.csv".format(directory),header=None,index_col=0)[1].sort_values()
```

```python
Carbon_color_dct = {k:c for k,c in zip(Carbon.index,colors)}
Enzyme_color_dct = {k:c for k,c in zip(Enzymes.index,["#737373"]*(len(Enzymes)-len(colors))+colors)}
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))
Carbon.plot.pie(autopct='%1.1f%%',ax=ax[0],
                             colors=[Carbon_color_dct[i] for i in Carbon.index])
Enzymes.plot.pie(autopct='%1.1f%%',ax=ax[1],
                              colors=[Enzyme_color_dct[i] for i in Enzymes.index])
plt.savefig("./figures/lethality-pies-aerobic.svg",format="SVG")
```

```python
pandas.Series({k:1 for k in Enzymes.index}).plot.bar(color=[Enzyme_color_dct[i] if i in Enzyme_color_dct else "#000000" for i in Enzymes.index],figsize=(8,0.3))
plt.savefig("./figures/lethality-pies-colorbar.svg",format="SVG")
```
