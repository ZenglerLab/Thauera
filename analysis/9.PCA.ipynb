```python
import pandas
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype'] = 'none'
```

```python
directory = "./thauera-files/ThaueraPaper/"
```

## Aerobic

```python
Carbon = pandas.read_csv("{}/carbonDetailsAer.csv".format(directory),header=None,index_col=0)[1]
lst = ["scoreCNEPSAer","scoreCNCarbonSourceAer","scoreYieldEPSAer","scoreYieldCarbonSourceAer"]
PCAS = {i:pandas.read_csv("{}/{}.csv".format(directory,i),header=None) for i in lst}
for k,df in PCAS.items():
    if "Carbon" in k:
        PCAS[k].index = Carbon.index
```

```python
colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00','#ffff33','#a65628']
carbon_color_dct = {i:c for i,c in zip(Carbon.unique(),colors)}
eps_color_dct = {i:c for i,c in zip(range(7),colors)}
```

```python
fig,ax = plt.subplots(2,2,figsize=(10,8))
ax=ax.flatten()
for idx,(k,pca) in enumerate(PCAS.items()):
    if "Carbon" in k:
        pca.plot.scatter(x=0,y=1,color=[carbon_color_dct[Carbon[i]] for i in pca.index],ax=ax[idx])
    else:
        pca.plot.scatter(x=0,y=1,color=[eps_color_dct[i] for i in pca.index],ax=ax[idx])
    ax[idx].set_title(k)
    ax[idx].set_xlabel("PC1")
    ax[idx].set_ylabel("PC2")
fig.tight_layout()
plt.savefig("./figures/pca-aerobic.svg",format="SVG")
```

```python
pandas.Series({k:1 for k in carbon_color_dct}).sort_index().plot.bar(color=[carbon_color_dct[i] for i in carbon_color_dct],figsize=(8,0.4))
plt.savefig("./figures/pca-aerobic-carbon-colorbar.svg",format="SVG")
```

```python
pandas.Series({k:1 for k in eps_color_dct}).sort_index().plot.bar(color=[eps_color_dct[i] for i in eps_color_dct],figsize=(8,0.4))
plt.savefig("./figures/pca-aerobic-eps-colorbar.svg",format="SVG")
```
